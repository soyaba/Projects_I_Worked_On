# Definició de grafs i digrafs

## Grafs

La llista d'adjacència és una de les maneres més eficients de definir un graf.

```python
llista_adj={0:[1,4,5],1:[0,2,6],2:[1,3,7],3:[2,4,8],4:[0,3,9],5:[0,7,8],6:[1,8,9],7:[2,5,9]}
llista_adj
```

```python
llista_adj2={0:[1,4,5],1:[2,6],2:[3,7],3:[4,8],4:[9],5:[7,8],6:[8,9],7:[9]}
llista_adj2
```

```python
G=Graph(llista_adj)
G
```

```python
G
```

```python
H=Graph(llista_adj2)
H
```

```python
G.is_isomorphic(H)
```

```python
G.adjacency_matrix()
```

```python
G.incidence_matrix()
```

```python
G.
```

Els grafs $G$ i $H$ són el mateix. A la llista *llista_adj* hi ha informació redundant.

## Digrafs

Encara que les dues llistes anteriors defineixen el mateix graf, no és així quan demanem digrafs. S'ha de tenir molt en compte aquest fet quan estem treballant amb digrafs.

```python
D=DiGraph(llista_adj)
D
```

```python
E=DiGraph(llista_adj2)
E
```

```python
D.is_isomorphic(E)
```

```python
D.adjacency_matrix()
```

```python
E.adjacency_matrix()
```

# Dibuixar grafs

Dues còpies mateix graf, dibuixades de dues maneres diferents, poden semblar dos grafs diferents. Haurem d'anar alerta i tenir en compte aquest fet quan dibuixem grafs.

Quan dibuixem diverses vegades un mateix graf, amb la mateixa comanda sense imposar opcions, ens poden sortir gràfics d'aparença diferent.

```python
dibuix=G.plot()
dibuix
```

```python
save(dibuix,"./G1.pdf")
```

```python
G.show()
```

En realitat, la comanda *show()* dóna coordenades al atzar als vèrtexs. Per aquesta raó, cada cop que la cridem el dibuix canvia.

Per evitar això, existeixen opcions que forcen l'aspecte del dibuix no variar gaire.

Una opció més utilitzada en aquest sentit és *layout* (consulteu l'ajut que dóna el sistema d'aquesta opció).

```python
dibuix=G.plot(layout='circular')
dibuix
```

```python
save(dibuix,"./G-circ.pdf")
```

```python
G.plot(layout='spring')
```

Podem donar la posició geomètrica dels vèrtexs. Així, quan dibuixem el graf, tindrem el gràfic exactament com volem.

```python
posicio={0: (0.0, 1.0),
       1: (-0.95, 0.30),
       2: (-0.58, -0.80),
       3: (0.58, -0.80),
       4: (0.95, 0.30),
       5: (0.0, 0.5),
       6: (-0.47, 0.15),
       7: (-0.29, -0.40),
       8: (0.29, -0.40),
       9: (0.47, 0.15)}
```

```python
dibuix=G.plot(pos=posicio)
dibuix
```

```python
save(dibuix,"./G-petersen.pdf")
```

### Producte cartesià ###

El producte cartesià de dos grafs, $G_1$ i $G_2$, denotat per $H=G_1\times G_2$, es defineix mitjançant el conjunt de vèrtexs

$$V(H)=V(G_1)\times V(G_2)$$

i conjunt de branques

$$E(H)=\{\{(a,b),(c,d)\}:~a=c\text{ i }\{b,d\}\in E(G_2)\text{ ò }\{a,c\}\in E(G_1)\text{ i }b=d\}$$

```python
graphics_array([graphs.CycleGraph(5).plot(),
graphs.CompleteGraph(2).plot()])
```

```python
G1=graphs.CycleGraph(5)
G2=graphs.CompleteGraph(2)
graphics_array([G1.plot(),G2.plot()])
```

```python
H=G1.cartesian_product(G2)
dibuix=H.plot()
dibuix
```

```python
save(dibuix,"./producte-cartesia.pdf")
```
